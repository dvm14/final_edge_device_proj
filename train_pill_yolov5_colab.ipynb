```
# 5. Train! 
# Update --data path below if data.yaml is in a different location
!python train.py \
  --img 640 \
  --batch 16 \
  --epochs 100 \
  --data ../Pill-Detection-2/data.yaml \
  --weights yolov5s.pt \
  --name pill_detection_colab \
  --cache \
  --workers 2
```
